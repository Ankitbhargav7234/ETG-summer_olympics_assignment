```python
import numpy as np
import pandas as pd
```

```python
ds=pd.read_csv('summer.csv')
```

1. In how many cities Summer Olympics is held so far?

```python
len(ds['City'].unique())
```

2. Which sport is having most number of Gold Medals so far?

```python
df=ds[ds['Medal']=='Gold']
data=[]
for i in df['Sport'].unique():
  data.append([i,len(df[df['Sport']==i])])
pd.DataFrame(data,columns=['Sport','Gold']).sort_values(by='Gold',ascending=False).head().plot(x='Sport',y='Gold',kind='bar',figsize=(15,5))
```

3. Which sport is having most number of medals so far? (Top 5)

```python
ds.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot.bar(figsize=(15,5))
```

4. Which player has won most number of medals? (Top 5)

```python
ds.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot.bar(figsize=(15,5))
```

5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df=ds[ds['Medal']=='Gold']
data=[]
for i in df['Athlete'].unique():
  data.append([i,len(df[df['Athlete']==i])])
pd.DataFrame(data,columns=['Athlete','Gold']).sort_values(by='Gold',ascending=False).head().plot(x='Athlete',y='Gold',kind='bar',figsize=(15,5))
```

6. In which year India won first Gold Medal in Summer Olympics?

```python
df[df['Country']=='IND']['Year'].min()
```

7. Which event is most popular in terms on number of players? (Top 5)

```python
ds.groupby('Event').count()['Athlete'].sort_values(ascending=False).head().plot.bar(figsize=(15,5))
```

8. Which sport is having most female Gold Medalists? (Top 5)

```python
df=ds[ds['Medal']=='Gold']
dj=df[df['Gender']=='Women']
bata=[]
for i in dj['Sport'].unique():
  bata.append([i,len(dj[dj['Sport']==i])])
pd.DataFrame(bata,columns=['Sport','Gold']).sort_values(by='Gold',ascending=False).head().plot(x='Sport',y='Gold',kind='bar',figsize=(15,5))
```
